```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, roc_curve
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
breast_cancer = load_breast_cancer()
df=pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
df['BreastCancer'] = breast_cancer.target
```

```python
data = load_breast_cancer()
X = data.data 
y = data.target 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Here we are doing some preprocessing to make it ready to do various classification algorithms. Which includes:
* Logistic Regression
* Decision Tree Classifier
* Random Forest Classifier
* Support Vector Machine
* K-Nearest Neighbours

# Logistic Regression

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression(random_state=42)
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
```

Logistic regression is a supervised machine learning algorithm that accomplishes binary classification tasks by predicting the probability of an outcome, event, or observation. Here it performed well mainly due to the dataset being simple.

# Decision Tree Classifier

```python
decision_tree_model = DecisionTreeClassifier(
    random_state=42,
    criterion="entropy",     
    max_depth=6,               
    min_samples_split=10,     
    min_samples_leaf=5         
)
decision_tree_model.fit(X_train, y_train)
```

```python
y_pred_tree = decision_tree_model.predict(X_test)
```

```python
def evaluate_model(y_true, y_pred):
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall:", recall_score(y_true, y_pred))
    print("F1 Score:", f1_score(y_true, y_pred))
```

```python
print("\nDecision Tree Performance:")
evaluate_model(y_test, y_pred_tree)
```

Decision Tree Classifier organize data into a tree-like structure where internal nodes represent decisions, branches represent outcomes and leaf node represent class labels.
